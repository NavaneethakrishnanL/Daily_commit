```python
def find_min_max(arr):
    if len(arr) == 0:
        return None, None  # Handle empty array case

    min_val = arr[0]
    max_val = arr[0]

    for num in arr:
        if num < min_val:
            min_val = num
        elif num > max_val:
            max_val = num

    return min_val, max_val
```

```python
arr = [12, 5, 8, 19, 3, 7]
min_val, max_val = find_min_max(arr)
print("Minimum:", min_val)
print("Maximum:", max_val)
```
