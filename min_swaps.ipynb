```python
def min_swaps(arr):
    n = len(arr)
    arr_pos = list(enumerate(arr))
    arr_pos.sort(key=lambda x: x[1])

    visited = [False] * n
    swaps = 0

    for i in range(n):
        if visited[i] or arr_pos[i][0] == i:
            continue

        cycle_size = 0
        j = i
        while not visited[j]:
            visited[j] = True
            j = arr_pos[j][0]
            cycle_size += 1

        swaps += cycle_size - 1

    return swaps
```

```python
arr = [4, 3, 2, 1]
```

```python
print(min_swaps(arr))  # Output: 2
```
