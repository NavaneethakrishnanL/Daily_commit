```python
numbers = [12, -7, 5, -3, 0, 9, -1]

def positive_negative_numbers(arr):
    positive_numbers = [num for num in arr if num >= 0]
    negative_numbers = [num for num in arr if num < 0]
    return positive_numbers, negative_numbers
```

```python
print("positive_negative_numbers:", positive_negative_numbers(numbers))
```
