```python
def word_frequency(tokens):
    freq = {}
    for t in tokens:
        t = t.lower()
        freq[t] = freq.get(t, 0) + 1
    return freq

text = "apple banana apple orange banana apple"
tokens = text.split()
print(word_frequency(tokens))
```
