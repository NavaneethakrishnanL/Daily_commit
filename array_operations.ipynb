```python
def reverse(arr):
    start=0
    end=len(arr)-1
    while start<end:
        arr[start],arr[end]=arr[end],arr[start]
        start+=1
        end-=1
    return arr
```

```python
print("\nArray_Elements_After_Custom_Reversal:", reverse(arr))
```

```python
arr=range(1,100)
def even_numbers(arr):
    even_numbers=[]
    for num in arr:
        if num%2==0:
            even_numbers.append(num)
    return even_numbers
```

```python
print("\nEven_Numbers_From_1_to_99:", even_numbers(arr))
```
